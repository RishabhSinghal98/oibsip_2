### Unemployment Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Dataset

```python
data_frame_1=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
data_frame_1.head()
```

```python
data_frame_2=pd.read_csv("Unemployment in India.csv")
```

```python
data_frame_2.head()
```

#### Copying Dataset 2 into new dataframe

```python
df2=data_frame_2.copy()
```

```python
df2['Year']=df2[' Date'].str.split('-').str[2] #extracting year from date column
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.dropna(inplace=True) #dropping null values
```

```python
df2
```

```python
df2.drop(' Date',axis=1,inplace=True)
```

```python
df2
```

```python
df2['Year']=df2['Year'].astype(int)
```

#### Dataframe for records Where Year is 2020

```python
df_for_2020=df2.loc[(df2['Year']==2020),['Region',' Frequency',' Estimated Unemployment Rate (%)',' Estimated Employed',' Estimated Labour Participation Rate (%)','Area','Year']]
```

#### Dataframe for records Where Year is 2019

```python
df_for_2019=df2.loc[(df2['Year']==2019),['Region',' Frequency',' Estimated Unemployment Rate (%)',' Estimated Employed',' Estimated Labour Participation Rate (%)','Area','Year']]
```

```python
df_for_2019.reset_index()
```

```python
df_for_2020.reset_index()
```

```python
df_for_2019
```

```python
df_for_2020
```

#### Conacatenate Datasets for Year 2020

```python
df_new=pd.concat([data_frame_1,df_for_2020],ignore_index=True) # dataframe which have data for year 2020
```

### Renaming the Columns

### Visualization For Year 2022

```python
plt.title("India Unemployment")
sns.histplot(data=df_new,x="Estimated Employed",hue="Region")

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("India Unemployment")
sns.histplot(data=df_new,x="Estimated Unemployment Rate",hue="Region")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("India Unemployment")
sns.histplot(data=df_new,x="Estimated Labour Participation Rate",hue="Region")
plt.show()
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(6,4))
sns.heatmap(df_new.corr(),annot=True)
plt.show()
```
